# Mongolia's Parliament is Aging

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We will import the parliament dataset the was downloaded from 1212.mn. Link to table is here: http://1212.mn/tables.aspx?TBL_ID=DT_NSO_3300_021V1

This dataset includes totals for each year and age brackets for all and also for females.

```python
parliament = pd.read_csv("Parliament-age-longform.csv")
```

```python
parliament.head()
```

```python
parliament_hist = parliament[(parliament.Indicator == '25-35') | (parliament.Indicator == '36-45') | (parliament.Indicator == '46+')].reset_index(drop=True)
```

```python
parliament_hist.head()
```

```python
# directory C:\\Users\\rober\\.matplotlib\\mpl_configdir\\stylelib\\test.mplstyle
plt.style.use('C:\\Users\\rober\\.matplotlib\\mpl_configdir\\stylelib\\mds.mplstyle')
```

```python
objects = parliament_hist.Indicator[parliament_hist.Year == 1992]
number = parliament_hist.Number[parliament_hist.Year == 1992]
plt.barh(objects, number, align='center', alpha=0.9)
plt.text(x = -4.1, y = 2.8, s = 'Parliament was relatively young in 1992', fontsize=26, weight='bold', alpha=.8)
plt.text(x = -4.1, y = 2.6, s= 'First parliament after democratic revolution', fontsize=18)
plt.text(x = -4.1, y = -0.95, s = '   ©Mongolian Data Stories                                                                       Source: National Statistical Office (1212.mn)   ', fontsize = 14, color = '#f0f0f0', backgroundcolor = 'dimgrey')
plt.xlabel('Number of Parliament Members')
plt.ylabel('Age Group')
plt.savefig('1992-parliament.png', dpi = 300, bbox_inches="tight")
```

```python
objects1 = parliament_hist.Indicator[parliament_hist.Year == 2016]
number1 = parliament_hist.Number[parliament_hist.Year == 2016]
plt.barh(objects1, number1, align='center', alpha=0.8)
plt.text(x = -4.1, y = 2.8, s = 'Almost no young blood', fontsize=26, weight='bold', alpha=.8)
plt.text(x = -4.1, y = 2.6, s= "Today's 2016 parliament seems to have aged without getting a new generation" , fontsize=18)
plt.xlabel('Number of Parliament Members')
plt.ylabel('Age Group')
plt.text(x = -4.1, y = -0.95, s = '   ©Mongolian Data Stories                                                                       Source: National Statistical Office (1212.mn)   ', fontsize = 14, color = '#f0f0f0', backgroundcolor = 'dimgrey')
plt.savefig('2016-parliament.png', dpi = 300, bbox_inches="tight")
```
